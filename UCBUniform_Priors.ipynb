```python
f_min=2e-5
f_max=1

def Prob_freq(f0):
    val=[]
    for i in (list(f0)):
        if (i>=10e-4) & (i<=10e-2):
            val.append((1/(10e-2-10e-4)))
        else :
            val.append(0)
    return val
    
f0=np.arange(f_min,f_max,10e-3)
plt.scatter(f0,Prob_freq((f0)))
plt.title('Frequency prior')
plt.xlabel(r'Frequency f0 (Hz)')
plt.ylabel(r'p($f_0$)')
plt.xlim([f_min,0.125])
print(f0,Prob_freq((f0)))

```

```python
f_min=2e-5
f_star=19.3e-3


def Prob_fder(f):
    val=[]
    for i in (list(f)):
        #der=
        a=-5*10e-6*i**(13/3)
        b=8*10e8*i**(11/3)
        if (i>=a) & (i<=b):
            val.append((1/(b-a)))
        else :
            val.append(0)
    print(val)
    return val
    

f_domain = np.linspace(f_min,f_star,10)
plt.title('Frequency derivative prior')
plt.xlabel(r'f (Hz)')
print(Prob_fder(f_domain))
plt.ylabel(r'p(df/dt)')
plt.plot(f_domain,Prob_fder(f_domain))
```

```python
def Prob_phi0(phi0):
    val=[]
    for i in (list(phi0)):
        if (i>=0) & (i<=2*np.pi):
            val.append((1/(2*np.pi)))
        else :
            val.append(0)
    return val
    
Phi_0=np.linspace(-0.5*np.pi,2.5*np.pi,10)
plt.scatter(Phi_0/np.pi,Prob_phi0((Phi_0)))
plt.title('Initial phase prior')
plt.xlabel(r'Angle (x $\pi$ rad)')
plt.ylabel(r'p($\chi_0$)')
print(Phi_0/np.pi,Prob_phi0(Phi_0))


#pol_angle
```

```python
def Prob_polAngle(phi0):
    val=[]
    for i in (list(phi0)):
        if (i>=0) & (i<=np.pi):
            val.append((1/(np.pi)))
        else :
            val.append(0)
    return val
    
polAngles=np.linspace(-0.5*np.pi,2*np.pi,10)
plt.scatter(polAngles,Prob_polAngle(polAngles))
plt.title('Polarization angle prior')
plt.xlabel(r'Angle (x $\pi$ rad)')
plt.ylabel(r'p($\psi $)')

print(polAngles/np.pi,Prob_polAngle(polAngles))

```

```python
def P_inclination(iVal):
    val=[]
    cos_val=[]
    for i in (list(iVal)):
        cos_val.append(np.cos(i))
    for i in (list(cos_val)):
        if (i>=-1) & (i<=1):
            val.append((1/(2*np.pi)))
        else :
            val.append(0)
    return val
    
incl_Val=np.linspace(-0.5*np.pi,2*np.pi,10)
plt.scatter(incl_Val,P_inclination(incl_Val))
plt.title('Inclination prior')
plt.xlabel(r'Angle (x $\pi$ rad)')
plt.ylabel(r'p($\kappa $)')

print(incl_Val/np.pi,P_inclination(incl_Val))
```

```python
from astropy.coordinates import SkyCoord
import astropy.units as u
import pandas as pd

alpha=0.25
Rb=0.8
Rd=2.5
Zd=0.4

def Prob_pos(beta,lamb):
    l=-lamb
    b=-beta       
    coord_gal=SkyCoord(l*u.rad, b*u.rad,frame="icrs")
    gc_icrs = coord_gal.transform_to('galactic')
    gc_l_rad,gc_b_rad=gc_icrs.l.radian,gc_icrs.b.radian
    coord_gal=SkyCoord(gc_l_rad*u.rad, gc_b_rad*u.rad,frame="galactic")
    coord_galc=coord_gal.cartesian
    xgc,ygc,zgc=pd.Series(np.array([(coord_galc.x),(coord_galc.y),(coord_galc.z)]))
    c = SkyCoord(x=xgc, y=ygc, z=zgc, representation_type='cartesian')
    x,y,z=pd.Series(np.array([(c.x),(c.y),(c.z)]))
            
    x=x-8.5
            
    r_sq=x**2+y**2+z**2
    u_sq=x**2+y**2

    return((1-0.1)*(alpha*np.exp(-r_sq/Rb**2)+(1-alpha)*np.exp(-np.sqrt(u_sq)/Rd)*(1/np.cosh((z/Zd)))**2)+0.1)
    
```
